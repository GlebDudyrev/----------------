## Бонус

На самом деле для решения этой задачи оптимальнее будет использовать запрос /timeframe, это уменьшит количество запросов и сократит время выполнения программы. Давайте посмотрим на возможный пример использования этого запроса.

```python
#Импорт библиотек
import requests
import json
import pandas as pd
```

```python
#Определяем временной промежуток
start_date = '2022-01-01'
end_date = '2022-01-07'

#Указываем адрес страницы к которой делаем запрос
url = "https://api.apilayer.com/currency_data/timeframe"
#Параметры запроса
params = {
    'start_date': start_date,
    'end_date': end_date
}
#Заголовок
headers= {
    "apikey": apikey
}
#Отправляем запрос
response = requests.request("GET", url, headers=headers, params=params)
status_code = response.status_code

#Производим десериализацию
result = json.loads(response.text)['quotes']
#Списки с датами и с котировками
date_list = []
qoutes_list = []

#Формируем списки
for date, quotes in result.items():
    date_list.append(date)
    qoutes_list.append(quotes)

#Получаем необхлдимую таблицу
data = pd.DataFrame(qoutes_list, index=date_list).iloc[:, :10]
display(data)
```
